#logistics regression - we find the probabilities of that record to be in a particular class
#most likely | Least likely - suppose we select standard line at 0.5, then we take probability values above 0.5 will be one class
#and below 0.5 will be another class

We fit a line like linear regression in logistic regression - that line gives you the value of probabilities

```python
import pandas as pd
logis_data = pd.read_csv('Logistic_reg_data.csv')
```

```python
logis_data
```

```python
X = logis_data.iloc[:,[2,3]].values
Y = logis_data.iloc[:,-1].values
```

```python
#splitting dataset
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.25, random_state = 121)
```

```python
#Normalize the data
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

```python
X_test
```

```python
#fitting logistics regression
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state=121)
classifier.fit(X_train,y_train)
```

```python
#Predict the test set results
y_pred = classifier.predict(X_test)
```

```python
#Making confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
cm
```

```python
import numpy as np
import matplotlib.pyplot as plt

#Visualizing the training set results
from matplotlib.colors import ListedColormap
X_set, y_set = X_train, y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))
#meshgrid creates pixel points, with start and stop range given and at step size of 0.01, which makes it seem continuous


plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
#contourf creates seperate sections for 2 classes

plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = ListedColormap(('red', 'green'))(i), label = j)
#The above loop creates red and green data points in their respective class in the graph

plt.title('Logistic Regression (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

```python
#Visualizing the test set results
from matplotlib.colors import ListedColormap
X_set, y_set = X_test, y_test
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))
#meshgrid creates pixel points, with start and stop range given and at step size of 0.01, which makes it seem continuous


plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
#contourf creates seperate sections for 2 classes

plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Logistic Regression (Test set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
