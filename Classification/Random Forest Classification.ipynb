#Ensemble learning - When you take multiple machine learning algorithms and put them together to make one big machine learing algorithm leveraging many algorithms

#Microsoft uses random forest in connect (live game) to understand how the body is moving and where in space the body part is. They were able to achieve faster speed and faster processing to reduce the hardware required. 

```python
import pandas as pd
random_forest_data = pd.read_csv('random_forest_data.csv')
random_forest_data
```

```python
X = random_forest_data.iloc[:,[2,3]].values
Y = random_forest_data.iloc[:,4].values
```

```python
#Splitting the dataset
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size = 0.2, random_state = 121)
```

```python
#feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
#fitting the classifier to the training set
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 10,criterion = 'entropy', random_state = 121) 
#The higher the enropy is the more your points are in disorder. 
#Ideally we want homogenous nodes with respect to class and hence we want less entropy
#Information gain is how much we managed to reduce the entropy from before and after the split: 
#entropy(parent node) - entropy(child node)

classifier.fit(X_train, y_train)
```

```python
#predicting the test set results
y_pred = classifier.predict(X_test)
```

```python
#Making confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
import numpy as np
import matplotlib.pyplot as plt

#Visualizing the training set results
from matplotlib.colors import ListedColormap
X_set, y_set = X_train, y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))
#meshgrid creates pixel points, with start and stop range given and at step size of 0.01, which makes it seem continuous


plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
#contourf creates seperate sections for 2 classes

plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = ListedColormap(('red', 'green'))(i), label = j)
#The above loop creates red and green data points in their respective class in the graph

plt.title('Random Forest (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

```python
#Visualizing the test set results
from matplotlib.colors import ListedColormap
X_set, y_set = X_test, y_test
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))
#meshgrid creates pixel points, with start and stop range given and at step size of 0.01, which makes it seem continuous


plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
#contourf creates seperate sections for 2 classes

plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Random Forest (Test set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
