```python
import pandas as pd
```

```python
data_linear_reg = pd.read_csv('Simple_linear_reg_data.csv')
```

```python
X = data_linear_reg.iloc[:,:-1].values
```

```python
Y = data_linear_reg.iloc[:,-1].values 
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 1/3, random_state = 121)
```

```python
#Fitting Simple Linear Regression model to Training set
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(X_train, y_train)
```

```python
#Predict on test set
```

```python
y_pred = regressor.predict(X_test)
```

```python
#Visualize the training set results
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue') #Plotting predicted salaries on training set on the same graph
plt.title('Salary vs Experience (Training Set)')
plt.xlabel('Years of Experience')
plt.ylabel('salary')
plt.show
```

```python
#Visualize the test set
plt.scatter(X_test, y_test, color = 'red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue') 
plt.title('Salary vs Experience (Training Set)')
plt.xlabel('Years of Experience')
plt.ylabel('salary')
plt.show
#Notice that the blue line here is the predicted values line on train set
```
